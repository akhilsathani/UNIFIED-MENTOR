# Bivariate Analysis

```python
#Let's check the relationship between each column
sns.scatterplot(x=df['sepal_length'],y=df['sepal_width'],data=df,hue=df['species'])
plt.show()
```

```python
sns.scatterplot(x=df['sepal_length'],y=df['petal_length'],data=df,hue=df['species'])
plt.show()
```

```python
sns.scatterplot(x=df['sepal_length'],y=df['petal_width'],data=df,hue=df['species'])
plt.show()
```

```python
sns.scatterplot(x=df['sepal_width'],y=df['petal_length'],data=df,hue=df['species'])
plt.show()
```

```python
sns.scatterplot(x=df['sepal_width'],y=df['petal_width'],data=df,hue=df['species'])
plt.show()
```

```python
sns.scatterplot(x=df['petal_width'],y=df['petal_length'],data=df,hue=df['species'])
plt.show()
```

```python
#finding out the correlation between each column
plt.figure(figsize=(10,8))
sns.heatmap(data=df.iloc[:,0:4].corr(),annot=True,cmap="YlGnBu")
plt.show()
```

# from the above it can be concluded that the sepal_length is positively correlated with petal_length and petal_width but negatively correlated with sepal_width and vice-versa

# Overall, the EDA of the Iris dataset reveals that features like petal length and petal width are highly discriminative for distinguishing between the three Iris species.
